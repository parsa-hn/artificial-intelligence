### Part 2
In this part, we should find teta0 and teta1 which would minimize the MSE function.<br>
In statistics, the mean squared error (MSE) of an estimator (of a procedure for estimating an unobserved quantity) measures the average of the squares of the errors â€” that is, the average squared difference between the estimated values and what is estimated. MSE is a risk function, corresponding to the expected value of the squared error loss. The fact that MSE is almost always strictly positive (and not zero) is because of randomness or because the estimator does not account for information that could produce a more accurate estimate. (From Wikipedia) <br>
In order to minimize the error and MSE function, partial derivate of the function with respect to the variables teta1 and teta0 should be calculated and be set to 0. This will make a system of linear equations with two variables and two equations. The formula to calculate these parameters are calculated as mentioned in [This Link](https://www.amherst.edu/system/files/media/1287/SLR_Leastsquares.pdf).<br>
After teta0 and teta1 parameters was calculated, we use the same data frame to determine whether the trainig was successful or not; So we plot a line using hypothesis function which is in red.<br>
As seen in the output of the code, MSE error is __0.002__ which is less than 0.1 thas is wanted in this part.

```python
def hypothesis(x, teta0, teta1):
    return teta0 + x*teta1

def MSE(x, y, teta0, teta1):
    divisor = len(x)*2
    sigmaArr = hypothesis(x, teta0, teta1) - y
    sigmaArrSqr = np.multiply(sigmaArr, sigmaArr)
    return np.sum(sigmaArrSqr) / divisor

x = attFrame['CGPA'].values
y = attFrame['Chance of Admit'].values
sigma1 = x - x.mean()
sigma2 = y - y.mean()

multDividend = np.multiply(sigma1, sigma2)
dividend = np.sum(multDividend)

multDivisor = np.multiply(sigma1, sigma1)
divisor = np.sum(multDivisor)

teta1 = dividend/divisor
teta0 = y.mean() - teta1*x.mean()

plt.plot(x, hypothesis(x, teta0, teta1), 'r')
plt.scatter(filledData["CGPA"], data["Chance of Admit"], label=col)
plt.legend()
print("MSE error : ", MSE(x, y, teta0, teta1))
```
